# Start cleaning data

```python
import pandas as pd
import calendar
df = pd.read_csv("../Initial Data/Madrid Daily Weather 1997-2015.csv")
```

### Add column day, month, year

- Change date format  to yy/mm/dd

- Create year month day title (For easy to call)

```python
df['CET'] = pd.to_datetime(df['CET'], format='%Y-%m-%d')
df['Day'] = df['CET'].dt.day
df['Month'] = df['CET'].dt.month
df['Month'] = df['Month'].apply(lambda x: calendar.month_name[x])
df['Year'] = df['CET'].dt.year
df
```

- Export csv file that do not delete outliner 

```python
df.to_csv('../Coding/Madrid Daily Weather 1997-2015 clean data add column.csv', index = False)
```

### Outliner

- Delete Events that only happend once (happend 1 time is hard to calculate and prediction)

```python
#Delete Events once happend
delete_events = df[' Events'].value_counts() 
# filter events that happend grater than 1 time
df = df[df[' Events'].isin(delete_events[delete_events > 1].index)]
# check current events
df[' Events'].value_counts()
```

- Export csv file that delete outliner

```python
df.to_csv('../Coding/Madrid Daily Weather 1997-2015 clean data.csv', index = False)
```
